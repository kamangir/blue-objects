# 🌀 log_matrix

```python
from typing import List
import numpy as np
import matplotlib.pyplot as plt
import cv2

from blueflow import notebooks

from blue_options import fullname, string
from blue_objects import objects, mlflow, file
from blue_objects.logger.matrix import log_matrix
from blue_objects.graphics.gif import generate_animated_gif

from blueflow import NAME, VERSION
from blueflow.logger import logger

logger.info(f"{NAME}-{VERSION}.{fullname()}, built on {string.pretty_date()}")
```

```python
object_name = objects.unique_object("test-log_matrix")
```

---

## pattern (by [OpenAI](https://chatgpt.com/c/673fe549-24f4-8005-9a24-b16b63d359c6))

```python
image_size = 200  
frequency = 5     
amplitude = 300   
offset = -20

x = np.linspace(0, 2 * np.pi * frequency, image_size)
y = np.sin(x) * amplitude + offset
matrix = np.tile(y, (image_size, 1))

plt.imshow(matrix, cmap='gray')

assert file.save_image(
    objects.path_of("pattern.png",object_name,),
    matrix.astype(np.uint8),
    log=True,
)
```

## log_matrix

```python
list_of_images: List[str] = []
for dynamic_range in range(50,500,50):
    filename: str = objects.path_of(
            filename=f"dynamic_range-{dynamic_range:03d}.png",
            object_name=object_name,
        )

    assert log_matrix(
        matrix=matrix,
        header=[],
        footer=[],
        filename=filename,
        dynamic_range=(-dynamic_range, dynamic_range),
        colormap= cv2.COLORMAP_JET,
        log_range=True,
    )

    list_of_images += [filename]
```

## gif

```python
gif_filename = objects.path_of("log_matrix.gif",object_name)

assert generate_animated_gif(
    list_of_images,
    gif_filename,
)

notebooks.imshow(gif_filename)
```
